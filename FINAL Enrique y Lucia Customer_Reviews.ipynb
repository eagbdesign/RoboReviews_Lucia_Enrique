```python
import nltk
from nltk.corpus import stopwords

# Descargar stopwords (solo la primera vez)
nltk.download('stopwords')

# Obtener la lista de stopwords en inglÃ©s
stop_words = set(stopwords.words('english'))

# FunciÃ³n para eliminar stopwords
def remove_stopwords(text):
    words = text.split()  # Dividir el texto en palabras
    filtered_words = [word for word in words if word not in stop_words]  # Eliminar stopwords
    return ' '.join(filtered_words)  # Volver a unir las palabras

# Rellenar valores nulos en la columna 'reviews.text'
df_combined["reviews.text"] = df_combined["reviews.text"].fillna("")

# Aplicar la funciÃ³n a la columna 'reviews.text'
df_combined["reviews.text"] = df_combined["reviews.text"].apply(remove_stopwords)
```

### Split Data into Training and Testing Sets

```python
x = df_combined["reviews.text"].map(lambda text: " ".join(text))
y = df_combined["reviews.rating"]
```

```python
ratings_to_class = {
    1: 0, 2: 0, 3: 1, 4: 2, 5: 2
}

y = y.map(ratings_to_class)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    x, 
    y, 
    test_size=0.2,     # proporciÃ³n de datos a usar para pruebas
    random_state=42    # asegura la reproducibilidad
)
```

### Train a Model and Fine - Tuning

```python
import numpy as np

# 1. Sklearn imports
from sklearn.model_selection import GridSearchCV
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.linear_model import LogisticRegressionCV
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 1. Crear el Pipeline
pipeline = Pipeline([
    ('count', CountVectorizer()),
    ('rf', RandomForestClassifier(random_state=42))
])

# 2. Definir los parÃ¡metros para la bÃºsqueda
#    Ten en cuenta que los hiperparÃ¡metros del RandomForest deben ir con el prefijo 'rf__'
#    y si quisieras configurar parÃ¡metros del CountVectorizer, usarÃ­as 'count__'
param_grid = {
    'count__ngram_range': [(1,1)],  # Ejemplo de cÃ³mo ajustar parÃ¡metros del CountVectorizer
    'rf__n_estimators': [100],
    'rf__max_depth': [None],
    'rf__min_samples_split': [10],
    'rf__min_samples_leaf': [1],
    'rf__max_features': ['log2'],
    'rf__bootstrap': [False],
    'rf__class_weight': ['balanced']
}

# 3. Configurar el GridSearchCV
rf_cv = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=5,              # ValidaciÃ³n cruzada con 5 particiones
    scoring='accuracy',
    n_jobs=-1,
    verbose=1
)

# 4. Entrenar (el Pipeline se encarga de vectorizar en cada fold de la CV)
rf_cv.fit(X_train, y_train)

# 5. Imprimir los mejores parÃ¡metros encontrados
print("Best params:", rf_cv.best_params_)

# 6. Evaluar el modelo con los mejores parÃ¡metros
y_pred_rf = rf_cv.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Precision:", precision_score(y_test, y_pred_rf, average='macro'))
print("Recall:", recall_score(y_test, y_pred_rf, average='macro'))
print("F1:", f1_score(y_test, y_pred_rf, average='macro'))
```

```python
y_pred_proba = rf_cv.predict_proba(X_test)
```

### Paso 1: Clustering con K-Means

```python
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt

# VectorizaciÃ³n con TF-IDF para el clustering
tfidf_vectorizer = TfidfVectorizer(max_features=5000, stop_words='english')
X_tfidf = tfidf_vectorizer.fit_transform(x)

# Determinar el nÃºmero Ã³ptimo de clusters usando el mÃ©todo del codo
inertia = []
for n in range(1, 11):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(X_tfidf)
    inertia.append(kmeans.inertia_)

# Visualizar el mÃ©todo del codo
plt.plot(range(1, 11), inertia, marker='o')
plt.xlabel('NÃºmero de clusters')
plt.ylabel('Inercia')
plt.title('MÃ©todo del codo para determinar el nÃºmero Ã³ptimo de clusters')
plt.show()

# Ajustar K-Means con el nÃºmero Ã³ptimo de clusters (supongamos 3)
n_clusters = 6
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
kmeans.fit(X_tfidf)

# AÃ±adir etiquetas de cluster al DataFrame original
df_combined['cluster'] = kmeans.labels_

# Mostrar las primeras filas con sus clusters
print(df_combined[['reviews.text', 'cluster']].head())
```
