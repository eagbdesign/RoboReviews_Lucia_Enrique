```python
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from sklearn.model_selection import train_test_split
from datasets import Dataset
import pandas as pd
import torch

# 1. Cargar el dataset combinado
df = pd.read_csv('archivo_combinado.csv')

# 2. Preparar datos para análisis de sentimientos
# Usar 'reviews.text' como entrada y 'reviews.rating' como etiquetas
df = df[['reviews.text', 'reviews.rating']].dropna()  # Eliminar valores nulos

# Convertir las etiquetas en clases (1=Negativo, 2=Neutral, 3=Positivo, por ejemplo)
df['label'] = df['reviews.rating'].apply(lambda x: 0 if x < 3 else (1 if x == 3 else 2))  # Ajustar según la escala

# Dividir en conjuntos de entrenamiento y prueba
train_texts, val_texts, train_labels, val_labels = train_test_split(
    df['reviews.text'].tolist(),
    df['label'].tolist(),
    test_size=0.2,
    random_state=42
)
```

```python


# 3. Tokenizar los textos usando DistilBERT
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')

train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=128)
val_encodings = tokenizer(val_texts, truncation=True, padding=True, max_length=128)

# 4. Crear dataset con PyTorch
class SentimentDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: torch.tensor(val[idx]) for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx])
        return item

train_dataset = SentimentDataset(train_encodings, train_labels)
val_dataset = SentimentDataset(val_encodings, val_labels)

# 5. Cargar DistilBERT preentrenado para clasificación de secuencias
model = DistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased', num_labels=3)  # 3 clases: negativo, neutral, positivo

# 6. Configurar el entrenamiento con Hugging Face Trainer
training_args = TrainingArguments(
    output_dir='./results',          # Directorio para guardar resultados
    evaluation_strategy='epoch',    # Evaluar al final de cada época
    learning_rate=5e-5,             # Tasa de aprendizaje
    per_device_train_batch_size=16, # Tamaño del lote en entrenamiento
    per_device_eval_batch_size=16,  # Tamaño del lote en evaluación
    num_train_epochs=3,             # Número de épocas
    weight_decay=0.01,              # Decaimiento del peso
    logging_dir='./logs',           # Directorio para guardar logs
    logging_steps=10,               # Frecuencia de logging
    save_steps=1000,                # Frecuencia de guardado
    save_total_limit=2              # Guardar solo los 2 mejores modelos
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
)

# 7. Entrenar el modelo
trainer.train()

# 8. Evaluar el modelo
trainer.evaluate()
```
